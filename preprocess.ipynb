```python
# Write dependencies here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import argparse
import re
import glob
import subprocess
```

```python
# Read the CSV file
df = pd.read_csv('data/Eber-insights-cig.csv')

# Display the dataframe with all columns
pd.set_option('display.max_columns', None)
print(df.head())
```

```python
# Check for columns with only NaN in the dataframe and store them in a list
nan_columns = df.columns[df.isnull().all()].tolist()

# Display nan_columns
print(nan_columns)

# Display the updated dataframe
df
```

```python
# Output non-NaN values for all columns in `nan_columns`
for col in nan_columns:
    print(df[col].unique())
# df['Store Staff'].unique()
```

```python
# Drop NaN columns if necessary
if nan_columns:
    df = df.drop(nan_columns, axis=1)

print(df.head())
```

```python
# Print the number of rows that is NaN in the `Member Id` column
print(f"Number of rows that is NaN in the `Member Id` column:", df['Member Id'].isnull().sum())

# Drop the nan values in the `Member Id` column
df = df.dropna(subset=['Member Id'])

# Print the number of rows that is NaN in the `Name` column with description for the print statement
print(f"Number of rows that is NaN in the `Name` column:", df['Name'].isnull().sum())

# Drop the nan values in the `Name` column
df = df.dropna(subset=['Name'])
```

```python
# Check for columns with NaN values
nan_columns = df.columns[df.isnull().any()].tolist()

# Display unique values for columns with NaN values
for col in nan_columns:
    unique_values = df[col].unique()
    print(f"Column: {col}, Unique Values: {unique_values}")
```

```python
df['Store Staff'] = df['Store Staff'].fillna('Admin')
df['Sign up at'] = df['Sign up at'].fillna('Tug Gelato')
```

```python
# Print nan rows for `Last Visit`
print(df[df['Last Visit'].isnull()])

# Fill NaN values in `Last Visit` with `Member Since` values
df['Last Visit'] = df['Last Visit'].fillna(df['Member Since'])

# Check for nan values
nan_columns = df.columns[df.isnull().any()].tolist()
print(nan_columns)
```

```python
# Print duplicate rows in the dataframe if any exist
print(df[df.duplicated()])

# Convert "Enrolled At", "Member Since" and "Last Visit" columns to datetime
df['Enrolled At'] = pd.to_datetime(df['Enrolled At'])
df['Member Since'] = pd.to_datetime(df['Member Since'])
df['Last Visit'] = pd.to_datetime(df['Last Visit'])

# Print the df info
print(df.info())
```

```python
# Check if `Member Since` and `Enrolled At` are equal
print(df['Member Since'].equals(df['Enrolled At']))

# Drop the `Enrolled At` column
df = df.drop('Enrolled At', axis=1)
```
